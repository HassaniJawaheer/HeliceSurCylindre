# Modélisation mathématiques

Dans l'espace orthonormé direct $(O,{\vec  i},{\vec  j},{\vec  k})$, il existe deux hélices circulaires (helice dextre et helice senestre) d'axes $(O, {\vec z})$ de rayon $r = a$ et de pas $ 2\pi b$ dont les équations paramétriques rectangulaires: 
$$
\left\{
    \begin{array}{ll}
        x(t) = a\cos(t) \\
        y(t) = a\epsilon \sin(t) \\
        z(t) = bt
    \end{array}
\right.
$$
avec $\epsilon = -1$(senestre) et $\epsilon = 1$(dextre) 

Où en coordonnées cylindriques :
$$
\left\{
    \begin{array}{ll}
        r = a \\
        \theta = b\epsilon \\
        z = \theta 
    \end{array}
\right.
$$

On modelisera comme suit: 
$$
\left\{
    \begin{array}{ll}
        x(\theta) = r cos(\theta) \\
        y(\theta) = r sin(\theta) \\
        z(\theta) = b\theta
    \end{array}
\right.
$$
avec $\epsilon = 1$ pour modéliser une helice dextre

# Initialisation des paramètres:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#Le rayon de l'helice et du cylindre:
r = 2
#le paramètre b du pas de l'helice et le pas de l'helice:
b = 1
# nombre d'arc de l'heclice :
nb_arc = 8
pas_helice = 2*b*np.pi
#l'angle maximale: nb_arc fois un tour de cercle:
theta_max = nb_arc*b*2*np.pi
#La hauteur du cylindre: qui doit égale à z_max c'est theta _max
height_cy = theta_max
```

# Modélisation :

```python
#le nombre de points:
n = 1000
theta = np.linspace(0, theta_max, n)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# La surface du cylindre:
z_cy = np.linspace(0, height_cy, n)
theta_grid, z_grid=np.meshgrid(theta, z_cy)
x_grid = r*np.cos(theta_grid)
y_grid = r*np.sin(theta_grid)
#alpha pour la transparence du cylindre:
ax.plot_surface(x_grid, y_grid, z_grid, alpha=0.1)
#Ajout de l'helice:
x = r*np.sin(theta)
z =  theta
y =  r*np.cos(theta)
ax.plot(x, y, z, 'r', lw=2, alpha = 1)
# On enlève les axes(raison esthétique)
ax.set_axis_off()
plt.show()
```
